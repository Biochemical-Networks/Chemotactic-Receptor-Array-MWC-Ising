## Ensemble averaging

The goal is to ensemble-average many realisations of the Gillespie algorithm. The problem is that the time steps differ between realisations. Here, we enforce regular time steps as follows. We determine the time points of the regular steps. Between those steps, we let the process run as usually, until it has reached another regular time point. Then, we average the variables over the interval between the two regular time points. In a sense, we can think of this as coarse-graining or pooling our variables.

```python
@numba.jit(nopython=True)

def gillespie_numba(a0,l0,m0,tf,tsteps,tchange,cnew,
                      E0=3,delta_f=-2,taul=0.01,Ki=18,Ka=2900,N=6,kR=0.1,kB=0.2,mMax=24,kAct=0.015,
                       methylation=True,ligand=True,activation=True):
    # note to self: parameters in units sec
    # taul = 0.01; kR = 0.1; kB = 0.2; kAct = 0.015
    '''
    Returns an array of the activity, number of ligands bound and methylation level
    for a rate-based MWC model including (de)activation, ligand (un)binding and
    (de)methylation, implemented using Gillespie's direct algorithm and Numba
    
    Parameters
    ----------
    a0: float
        initial activity
    l0: float
        initial number of ligands bound
    m0: float
        initial methylation level
    tf: float
        final time of simulation in sec
    tsteps: int
        number of evenly-spaced time steps
    tchange: float
        time in sec at which the concentration is changed
    cnew: float
        new concentration in micromolar
    E0: float
        free energy of unmethylated active cluster in units of kT
    delta_f: float
        energy penalty per methylated site in units of kT
    taul: float
        ligand (un)binding time scale
    Ki: float
        dissocation constant for inactive cluster in micromolar
    Ka: float
        dissociation constant for active cluster in micromolar
    N: int
        number of receptors per cluster
    kR: float
        methylation rate per sec
    kB: float
        demethylation rate per sec
    mMax: int
        number of methylation sites (note: mMax = 4*N)
    kAct: float
        activation rate per sec
    methylation: bool
        boolean for whether methylation is taken into account
    ligand: bool
        boolean for whether ligand (un)binding is taken into account
    activation: bool
        boolean for whether (de)activation is taken into account
        
    Returns
    -------
    av_states: NumPy array
        array of activity, number of ligands bound and methylation level, averaged between
        the time steps given in time_regular
    '''
    # set the initial state
    state = np.array([a0,l0,m0],dtype='float64')
    
    time_regular = np.linspace(0,tf,num=tsteps)
    
    time = 0
    c = 100
        
    # set rate constant for ligand binding
    ######## change Ka here?
    kplus = 1/(taul*(c + Ka))
    
    for i in range(tsteps):
        # counter for time steps
        j = 0
        
        while time <= time_regular[i]:
            # set the events with the corresponding rates
            rates = np.zeros(4)
            events = np.zeros((4,3))
            
            # change concentration after tchange
            if time > tchange:
                c = cnew
                
            # calculate free-energy difference
            f = E0 + state[1]*np.log(Ka/Ki) + delta_f*state[2]
            
            # set rate constants for (de)activation
            kDeact = kAct*np.exp(f)
                        
            # check methylation states
            meth_ceiling = state[2] < mMax
            meth_floor = state[2] > 0
            
            # for inactive cluster
            if state[0] == 0:
                # ligand binding
                rates[0] = kplus*c*(N - state[1])*ligand
                events[0] = [0,1,0]
                
                # ligand unbinding
                rates[1] = kplus*Ki*state[1]*ligand
                events[1] = [0,-1,0]

                # methylation; only possible if a site is available
                rates[2] = kR*meth_ceiling*methylation
                events[2] = [0,0,1]

                # activity switching
                rates[3] = kAct*activation
                events[3] = [1,0,0]

            # for active cluster
            else:
                # ligand binding
                rates[0] = kplus*c*(N - state[1])*ligand
                events[0] = [0,1,0]
                
                # ligand unbinding
                rates[1] = kplus*Ka*state[1]*ligand
                events[1] = [0,-1,0]

                # demethylation; only possible if a site is occupied
                rates[2] = kB*meth_floor*methylation
                events[2] = [0,0,-1]

                # deactivation
                rates[3] = kDeact*activation
                events[3] = [-1,0,0]

            # this samples the time until the next event
            u = np.random.uniform(0,1)
            dt = -np.log(u)/np.sum(rates)

            # update time
            time += dt

            # determine which event occurs
            p = np.random.uniform(0,1)*np.sum(rates)
            mask = p < np.cumsum(rates)
            event_idx = np.where(mask == True)[0][0]
            
            # perform the event
            state = state + events[event_idx]
            
            if j == 0:
                all_states = state
            else:
                all_states = np.append(all_states,state)
            
            j += 1
            
        # reshape into a 2d array
        reshaped_states = np.reshape(all_states,(all_states.size//3,3))
        
        # averaging (the weird numba way)
        av_states = np.zeros(3)
        for i in range(3):
            # prevent a numba ZeroDivison error
            if reshaped_states[:,i].size > 0:
                av_states[i] = np.mean(reshaped_states[:,i])
            else: 
                av_states[i] = 0

        # store all results
        if i == 0:
            results = av_states
        else:
            results = np.append(results,av_states)
                
    # reshape into a 2d array
    reshaped_results = np.reshape(results,(results.size//3,3))
                                    
    return reshaped_results

@numba.jit(nopython=True)

def repeated_gillespie(n_sims,tf,tsteps,tchange,cnew,meth_bool=True,lig_bool=True,act_bool=True,
                       a0=0,l0=0,m0=0):
    # array to store results; first index is the realisation, second is time
    # third specifies the variable (e.g., activity)
    all_sims = np.empty((n_sims, tsteps, 3))

    for i in range(n_sims):
        all_sims[i, :, :] = gillespie_numba(a0,l0,m0,tf,tsteps,tchange,cnew,
                                            methylation=meth_bool,ligand=lig_bool,activation=act_bool)
        
    return all_sims
```

## Dose-response Curve

In this section, I use the simulation to extract the dose-response curve. The procedure is as follows. After simulating the system, I let it equilibrate until the relative change between successive time points is small. This allows to compute a baseline. Subsequently, the cell receives a step change in concentration (for now, I focus on an increase in concentration). In this case, I focus on the amplitude of the response, which corresponds to the dip after the change in concentration relative to the baseline activity. Repeating this procedure for various values of the concentration increase allows to compute a dose-response curve.

Note that since we are only interested in the response right after the change in concentration, we can speed up the computation by making the simulation time shorter.

```python
def baseline(activity,epsilon,tchange_idx):
    # compute the relative change between successive steps
    # note a small numerical term to prevent division by zero
    relative_change = np.diff(activity)/(activity[1:] + 10**-6)
    
    # check where relative change is small enough for the first time
    eql_start = np.where(np.abs(relative_change) < epsilon)[0][0]
    
    # average activity during equilibrium
    a_base = np.mean(activity[eql_start:tchange_idx])
    
    return a_base, eql_start
```

```python
def response_finder(activity,a_base,eql_start,addition):
    # remove transient
    eql_act = activity[eql_start:]
    
    # find absolute response depending on whether the concentration is increased or decreased
    if addition:
        response = np.min(eql_act - a_base)
    else:
        response = np.max(eql_act - a_base)
    
    return response
```

```python
def dose_response_finder(c_delta_arr,n_sims,tf,tsteps,tchange,epsilon):
    # array to save results; first entry is removed later on
    all_responses = np.array([0.])
    
    # find index at which concentration changes
    tchange_idx = int(np.floor(tsteps*tchange/tf))
    
    for i,c_delta in enumerate(c_delta_arr):
        # change the concentration
        cnew = 100 + c_delta
        
        # obtain arrays of activity (dumb numba way)
        results = repeated_gillespie(n_sims,tf,tsteps,tchange,cnew)
        
        activity = np.mean(results[:,:,0], axis=0)
        # the commented section is to be used with numba
        #activity = np.zeros(tsteps)
        
        #for step in range(tsteps):
        #    if results[:,step,0].size > 0:
        #        activity[step] = np.mean(results[:,step,0])
            
        # determine the response
        a_base, eql_start = baseline(activity,epsilon,tchange_idx)
        
        if c_delta >= 0:
            addition = True
        else:
            addition = False
            
        response = response_finder(activity,a_base,eql_start,addition)
        
        # save result
        all_responses = np.append(all_responses,response)
        if (i+1) % 5 == 0:
            print(100*(1+i)/c_delta_array.size,'% done')
        
    return all_responses[1:] # remove first entry 
```

```python
epsilon = 10**-2
n_sims = 10**4
tf = 500
tsteps = 10**2
tchange = 200

c_delta_array = -np.logspace(0,2,5)

responses = dose_response_finder(c_delta_array,n_sims,tf,tsteps,tchange,epsilon)
```

```python
plt.semilogx(-c_delta_array, np.abs(responses),'.')
plt.xlabel(r'$\Delta c$')
plt.ylabel('Response')
plt.show()
```

## Fitting to Extract Response and Adaptation Timescale

This part uses a model proposed by Mattingly et al. (2021) for the shape of the response curve as

$$ K(t) = G \exp\{-t/\tau_2\}(1 - \exp\{-t/\tau_1\})H(t), $$

where $K(t)$ is the activity, $G$ is the gain or amplitude, $\tau_1$ is the rise time, $\tau_2$ is the adaptation time and $H(t)$ is the Heaviside step function. Note that we have subtracted the baseline activity and set $t=0$ when the concentration is increased. Furthermore, Mattingly et al. only probed responses for a 10% increase or decrease in concentration, starting from a background of 100 $\mu$M.

```python
# obtain nice data

n_sims = 10**5
tf = 500
tsteps = 10**2
tchange = 200
cnew = 90

time = np.linspace(0,tf,num=tsteps)

data = repeated_gillespie(n_sims,tf,tsteps,tchange,cnew)

activity = np.mean(data[:,:,0],axis=0)

plt.plot(time,activity)
plt.show()
```

```python
# shifting the time and activity
epsilon = 10**-2
tchange_idx = int(np.floor(tsteps*tchange/tf))
shifted_activity = activity[tchange_idx:] - baseline(activity,epsilon,tchange_idx)[0]

# shift time such that tchange=0
mask = (time - tchange) >= 0
shifted_time = time[mask] - tchange

# check the result
plt.plot(shifted_time,shifted_activity)
plt.show()
```

```python
# define model function
def mattingly_fit(t,G,tau1,tau2,alpha):
    return G*np.exp(-t/tau2)*(1 - np.exp(-t/tau1)) + alpha

# perform fitting
########## currently I have added a constant to the model instead of shifting the activity
popt, pcov = curve_fit(mattingly_fit,shifted_time,activity[tchange_idx:])
print('Parameters of best fit: [G, tau1, tau2, alpha] =',popt)

plt.plot(shifted_time,activity[tchange_idx:],label='Simulation')
plt.plot(shifted_time,mattingly_fit(shifted_time,*popt),label='Fit')
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Shifted activity')
plt.show()
```
