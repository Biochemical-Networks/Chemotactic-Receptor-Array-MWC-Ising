```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# check that I can export data and show figures

def test_func(arr):
    return arr + np.random.uniform(-2,2,arr.size)

arr = np.linspace(10,20,5)

np.savetxt('test_output.txt',test_func(arr))

plt.plot(arr,test_func(arr),'o')
plt.plot(arr,arr,'r:')
plt.show()
```
